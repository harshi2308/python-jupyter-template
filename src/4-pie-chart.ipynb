### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

# Sample data
labels = ['Apple', 'Banana', 'Orange', 'Grapes']
sizes = [15, 30, 45, 10]

# Create a pie plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'yellow', 'orange', 'purple'])

# Adding title
plt.title("Fruit Distribution")

# Display the plot
plt.show()
```
