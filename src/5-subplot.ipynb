### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2)

# Data for the plots
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot 1: Line plot
axes[0, 0].plot(x, y, color='blue')
axes[0, 0].set_title("Line Plot")

# Plot 2: Bar plot
axes[0, 1].bar(x, y, color='green')
axes[0, 1].set_title("Bar Plot")

# Plot 3: Scatter plot
axes[1, 0].scatter(x, y, color='red')
axes[1, 0].set_title("Scatter Plot")

# Plot 4: Pie plot
sizes = [15, 30, 45, 10]
labels = ['Apple', 'Banana', 'Orange', 'Grapes']
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title("Pie Plot")

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
